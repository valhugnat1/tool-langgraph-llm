```python
# rag.py file

from dotenv import load_dotenv
import psycopg2
import os

# Load environment variables
load_dotenv()

# Establish connection to PostgreSQL database using environment variables
conn = psycopg2.connect(
    database=os.getenv("PGDATABASE"),
    user=os.getenv("PGUSER"),
    password=os.getenv("PGPASSWORD"),
    host=os.getenv("PGHOST"),
    port=os.getenv("PGPORT"),
)


```

```python
try:
    with conn.cursor() as cur:
        cur.execute("SELECT 1")
    print({"status": "healthy", "database": "connected"})
except Exception as e:
    print({"status": "unhealthy", "database": str(e)})
```

```python
# Create a cursor to execute SQL commands
cur = conn.cursor()

# Execute a simple SQL command to check the connection
cur.execute("SELECT object_key, metadata FROM object_loaded;")

# Fetch all the results
rows = cur.fetchall()

i = 0
# Print the results
for row in rows:
    print(i, row)
    i += 1
```

```python
import json

# Function to edit the metadata column for a specific row (filtered by object_key)
def edit_metadata(object_key, new_metadata):

    new_metadata_json = json.dumps(new_metadata)
    # Create a cursor to execute SQL commands
    cur = conn.cursor()
    # SQL command to update the metadata column
    update_query = """
        UPDATE object_loaded 
        SET metadata = %s 
        WHERE object_key = %s;
    """
    # Execute the update query with the new metadata and specified object_key
    cur.execute(update_query, (new_metadata_json, object_key))
    # Commit the changes to the database
    conn.commit()
    # Close the cursor
    cur.close()
```
